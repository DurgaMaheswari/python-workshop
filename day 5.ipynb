```python
#"." symbol
print(re.search("..","vignan"))
print(re.search("..","v"))
```

```python
#{min,max}
print(re.search(".{3,10}","vignan lara"))
print(re.search("..","v"))
print(re.search(".{0,12}","v"))
print(re.search(".{10,12}","vignan lara"))
```

```python
#"^"
print(re.search("^vignan","vignan lara"))
print(re.search("^vignan","nan lara"))
print(re.search("vignan","ghfvignan lara"))
```

```python
#"$"
print(re.search("ra$","vignan lara"))
```

```python
# "*"
print(re.search(" *","vignan lara"))
print(re.search(" ","vignan lara"))
print(re.search(" ","vignanlara"))
```

```python
#"+"
print(re.search(" +","vignan lara"))
print(re.search(" +","vignan             lara"))
print(re.search("z+","vignan lara"))
```

```python
#"[]"
print(re.search("[VLITS]","Vignan lara"))
print(re.search("[VLITS]","ignan  Lara"))
print(re.search("[VLITS]","vignan lara"))
print(re.match("[VLITS]","Vignan Lara"))
print(re.match("[VLITS]","vignan  Lara"))
print(re.match("[VLITS]","vignan lara"))
```

```python
#"[^]"
print(re.search("[^VLITS]","Vignan lara"))
print(re.search("[^VLITS]","ignan  Lara"))
print(re.search("[^VLITS]","vignan lara"))
print(re.match("[^VLITS]","Vignan Lara"))
print(re.match("[^VLITS]","vignan  Lara"))
print(re.match("[^VLITS]","vignan lara"))
```

```python
#\d,\D,\s,\S
print(re.search("\d","vignan 123"))
print(re.search("\D","45vignan 123"))
print(re.search("\d\d\d","vignan 12356"))
print(re.search("\s","vignan 123"))
print(re.search("\S","   Vignan 123"))
```

# comprehension

```python
#list []
# syntax [output loop condition]
```

```python
l=[1,2,3,4,5,6]
for i in l:
    if i%2==0:
        print(i**2)
```

```python
print(l)
[i for i in l if(i%2==0)]
```

```python
print([i**2 for i in [1,2,3,5,6] if i%2==0])
```

```python
[True if(i%2==0) else False for i in l]
```

# dictionary comprehension
-{key:value loop condition}


```python
#{1:1,12:4,3:9}
d={}
for i in l:
    d[i]=i**2
print(d)
    
```

```python
x={i:i**2 for i in l}
x
```

```python
l=[1,1,2,3,4,4,5,6,2]
#{1:2,2:2,3:1,4:1,5:1,6:1}
d={}
for i in l:
    if i not in d:
        d[i]=i
    else:
        d[i]=d[i]+1
print(d)
```

```python
d={}
for i in l:
    d[i]=l.count(i)
print(d)
```

```python
{i:l.count(i) for i in l}
```

# oops concept in python
    -oops is all about creating an object
    -  an object is the instance of class we can say and it is a group of interelated varables and functions
    -variables are the properties of the class
    -functions are the behaviour of the class
    -feautures:properties
    -functio-work

```python
class car:
    def __init__(c,name,milege):
        c.name=name
        c.milege=milege
    def max_speed(c,speed):
        c.speed=speed
        return f"the {c.name} runs at max speed {speed}km/h" 
honda=car("honda",21.4)
print(honda.max_speed(150))
print()
```

```python
class car:
    def __init__(c,name,milege):
        c.name=name
        c.milege=milege
    def max_speed(c,speed):
        c.speed=speed
    def description(c):
        return f"the {c.name} runs at max speed {c.milege}km/h" 
class bmw(car):
    pass
b=bmw("bmw",35.3)
print(b.description())
```

```python
def second():
    yield 1
    yield 2
s=second()
for num in s:
    print(num,end=" ")
```
